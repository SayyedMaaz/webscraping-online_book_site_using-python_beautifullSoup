```python
import requests

from bs4 import BeautifulSoup


import pandas as pd
```

```python
web=requests.get("https://www.goodreads.com/list/show/47649.Best_Books_of_2014")
web
```

```python
sp=BeautifulSoup(web.content)
sp
sp.prettify
```

```python
title=sp.find_all("a",class_="bookTitle")
title
title1=[]
for i in title:
    title1.append(i.text.replace("\n",""))
title1
```

```python
author=sp.find_all("a",class_="authorName")
author
authors=[]
for i in author:
    authors.append(i.text.replace("\n",""))
authors
```

```python
score=sp.find_all("span",class_="smallText uitext")
score
scores=[]
for i in score:
    scores.append(i.text.replace(" ","").replace("\n","").replace("\u2003","").replace(",and"," "))
scores
```

```python
rating=sp.find_all("span",class_="minirating")
rating
ratings=[]
for i in rating:
    ratings.append(i.text.replace("\n","").replace("—",""))
ratings
```

```python
wish=sp.find_all("span",class_="progressTrigger")
wish
wishs=[]
for i in wish:
    wishs.append(i.text.replace("\n",""))
wishs
```

```python
giverate=sp.find_all("div",class_="myRating uitext greyText")
giverate
giverates=[]
for i in giverate:
     giverates.append(i.text.replace("\n",""))
giverates
```

```python
cover=sp.find_all("img",class_="bookCover")
cover
covers=[]
for i in cover:
     covers.append(i.img)
covers
```

```python
rank=sp.find_all("td",class_="number")
rank
ranks=[]
for i in rank:
     ranks.append(i.text.replace("\n",""))
ranks
```

```python
print(len(title1),len(scores),len(ratings),len(authors),len(ranks),len(wishs),len(giverates),len(cover))
```

```python

df=pd.DataFrame({})
df["TITLE NAME"]=title1
df["AUTHOR NAME"]=authors
df["RANK"]=ranks
df["RATINGS"]=ratings
df["SCORES"]=scores
df["WISH_LIST"]=wishs[0:100]
df["GIVE_RATES"]=giverates
df["cover"]=cover
df
```

```python
df.to_csv("file1.csv")
df
df=pd.read_csv("file1.csv")   
df    #file location same hai to path dene ki zarorat nhi warna opath bhii dena poadega
```
